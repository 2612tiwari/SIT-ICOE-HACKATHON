```python
import pandas as pd

# Read the CSV files
d1 = pd.read_csv("student-mat.csv", sep=";")
d2 = pd.read_csv("student-por.csv", sep=";")

# Merge the dataframes on specified columns
d3 = pd.merge(d1, d2, on=["school", "sex", "age", "address", "famsize", "Pstatus", "Medu", "Fedu", "Mjob", "Fjob", "reason", "nursery", "internet"])

print(len(d3))
print(d3.columns)
```

```python
print(d3.head())
```

```python
import pandas as pd

# Define the columns for grades in mathematics (G1_x, G2_x, G3_x) and Portuguese (G1_y, G2_y, G3_y)
math_columns = ['G1_x', 'G2_x', 'G3_x']
portuguese_columns = ['G1_y', 'G2_y', 'G3_y']

# Calculate the correlation matrix between the specified columns
correlation_matrix = d3[math_columns + portuguese_columns].corr(method='spearman')

print("Spearman correlation matrix between grades in mathematics and Portuguese:")
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming corr_matrix is your correlation matrix
# Visualize the correlation matrix using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```
